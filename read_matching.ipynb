# Read matching result

```python
import pandas as pd
import json
import os
```

```python
BASE_DIR = "output/matching"

def load_matching(name, base_dir=BASE_DIR):
    """
    Load the matching results from the specified folder.
    """
    folder = os.path.join(base_dir, name)
    
    matches = pd.read_parquet(os.path.join(folder, "matches.parquet"))
    profiles = pd.read_parquet(os.path.join(folder, "profiles.parquet"))
    balance = pd.read_parquet(os.path.join(folder, "balance.parquet"))
    
    with open(os.path.join(folder, "config.json"), "r") as f:
        config = json.load(f)
    
    return {
        "matches": matches,
        "profiles": profiles,
        "balance": balance,
        "config": config
    }
```

```python
res = load_matching("summary_season")

print(res["matches"])
print(res["profiles"])
print(res["config"])
```
